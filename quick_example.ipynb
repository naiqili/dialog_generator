The example of the dictionary (optional).

```python
import cPickle
from pprint import pprint

(ind2word, word2ind, ontology_bool, ontology_time, ontology_str) = cPickle.load(open('tmp/dict.pkl'))

print 'ind2word: a dict from index to word'
pprint(ind2word.items()[-5:])

print
print 'ind2word: a dict from word to index'
pprint(word2ind.items()[-5:])

print
print 'ontolog_bool: a list on the ontologies that are of bool type'
print(ontology_bool)

print
print 'ontolog_time: a list on the ontologies that are of integer type'
print(ontology_time)

print
print 'ontolog_str: a list on the ontologies that are of integer string'
print(ontology_str)
```
